### Libraries

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
```

### Reading the Image

```python
### YOUR CODE ###
img = cv2.imread("q1.png", 0)
```

### Finding Threshold

```python
# You can use cv2 methods for finding this
_, thresholded = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY)
```

### Finding Connected Components

```python
# You can use cv2 methods for finding them
### YOUR CODE ###
num_labels, labels = cv2.connectedComponents(thresholded)
```

#### Show the result

Show each component with a random color, and write the number of components on the result image

```python
### YOUR CODE ###
output = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
colors = np.random.randint(0, 255, size=(num_labels, 3), dtype=np.uint8)
colors[0] = [0, 0, 0]

for y in range(img.shape[0]):
    for x in range(img.shape[1]):
        label = labels[y, x]
        output[y, x] = colors[label]


cv2.putText(output, f"Components: {num_labels - 1}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 
            1, (255, 255, 255), 2)

output_rgb = cv2.cvtColor(output, cv2.COLOR_BGR2RGB)

plt.imshow(output_rgb)
plt.title("Connected Components")
plt.axis('off')
plt.show()
```
