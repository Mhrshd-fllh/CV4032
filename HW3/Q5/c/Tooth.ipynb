# Import Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

# Load Image

```python
# Path to the image
image_path = "Tooth.png"  # Set your image path here         #TO DO
```

# a) Read the image in grayscale

```python
image_gray = cv2.imread(image_path, 0)       #TO DO
```

# b) Apply pseudo-coloring using a suitable colormap

```python
image_color = cv2.applyColorMap(image_gray, cv2.COLORMAP_JET)     #TO DO
```

# c) Display both grayscale and pseudo-colored images

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title("Original Grayscale X-ray")
plt.imshow(image_gray, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Pseudo-colored X-ray (COLORMAP_JET)")
plt.imshow(cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.tight_layout()
plt.show()
```
