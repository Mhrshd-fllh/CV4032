```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("image.png")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
```

```python
increase = 64
brightness_rgb = image_rgb + increase
brightness_rgb[brightness_rgb > 255] = 255


plt.imshow(brightness_rgb)

```

```python
hls_image = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2HLS)
hls_image[:, :, 1] = hls_image[:, :, 1] + (increase * 255  / 255)
hls_image[:, :, 1][hls_image[:, :, 1] > 255] = 255

bright_hs1 = cv2.cvtColor(hls_image, cv2.COLOR_HLS2RGB)

plt.imshow(bright_hs1)
```
